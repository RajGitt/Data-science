```python
import pandas as pd
```

```python
d = {"Sales":[1000000,222000,10000000,522000,111111,222222,1111111,20000,75000,90000,1000000,10000], 
     "City":["Tampa","Tampa","Orlando","Jaksonville","Miami","Jaksonville","Miami","Miami","Orlando","Orlando","Orlando","Orlando"],
     "Size":["Small","Medium","Large","Large","Small","Medium","Large","Small","Medium","Medium","Medium","Small"]}
```

```python
#Creating a DataFrame
df = pd.DataFrame(data=d)
```

```python
df
```

```python
#Now we are converting the City column into numerical value as it has nominal categorical values, by OneHotEncoder. 
#And converting the Size column into numerical value(0,1,2) as it is has ordinal categorical values, by OrdinalEncoder.

#Here to transform the columns at once by using OrdinalEncoder and OneHotEncoder we will use the make_column_transformer.

#importing OrdinalEncoder and OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder , OneHotEncoder
oe = OrdinalEncoder()
#Here we'll use two parameters inside OneHotEncoder
#handle_unknown="ignore", if we have trained our data and will provide new data as an input then it can generate error message so handle_unknown="ignore" will ignore if there is unkown data.
#sparse_output=False, will return us the numpy dense array or pandas DataFrame but this takes more memory but it is more redable, And if the sparse_output=True then it will return us the sparse matrix as it return the data wit lots of zeros makeing it more memory efficient 
ohe = OneHotEncoder(handle_unknown="ignore",sparse_output=False)
```

```python
#Now we are importing make_column_transformer, so that we can use both the OrdinalEncoder and OneHotEncoder on the columns at once(in the same code of line).
from sklearn.compose import make_column_transformer
```

```python
#Instantiating the object.
mct = make_column_transformer(
    (ohe,["City"]),
    (oe,["Size"]),
    #here the remainder="passthrough" parmater is used because we are using these OrdinalEncoder and OneHotEncoder on two columns but there is one remainder column
    #where we are doing nothing so, we can simply pass it through without using any method. So, by using the remainder="passthrough" we have included the sales column
    #but we can drop it, if we want to by remainder="drop"
    remainder="passthrough"
).set_output(transform="pandas")#.set_output(transform="pandas") will convert the data into pandas DataFrame.
```

```python
#Now we will fit_transform these columns on which we have appied the make_column_transformer
mct.fit_transform(df)
```

```python
#2) Second way if we want to drop the sales columns then,
```

```python
mct2 = make_column_transformer(
    (ohe,["City"]),
    (oe,["Size"]),
    remainder="drop"
).set_output(transform="pandas")
```

```python
#Using the fit_transform to fit will take the column as an input and transform will transform the columns in integer value.
mct2.fit_transform(df)
```

```python
#if we want to use the class(OrdinalEncoder or OneHotEncoder),keep some of the columns and drop some of the columns, then how to do it
mct3 = make_column_transformer(
    (ohe,["City"]),#we are only using the OneHotEncoder on one column
    ("passthrough",["Size"]),#we can use passthrough like this as a step
    remainder="drop"
).set_output(transform="pandas")
```

```python
#Using the fit_transform to fit will take the column as an input and transform will transform the columns in integer value.
mct3.fit_transform(df)
```
