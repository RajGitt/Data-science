```python
#Import Libraries and Load Data

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load datasets
load_src = r"C:\Users\Hp\Downloads\Project\TATAMOTORS_BSE_01Oct20-01Oct21.csv"
bse_data = pd.read_csv(load_src)
load_src2 = r"C:\Users\Hp\Downloads\Project\TATAMTRDVR_NSE_01Oct20-01Oct21.csv"
nse_data = pd.read_csv(load_src2)
load_src3 = r"C:\Users\Hp\Downloads\Project\TTM_NYSE_01Oct20-01Oct21.csv"
nyse_data = pd.read_csv(load_src3)
```

```python
#Data Preprocessing
#We will clean and preprocess each dataset:

# Convert 'Date' columns to datetime

bse_data['Date'] = pd.to_datetime(bse_data['Date'], format='%Y-%m-%d')
nse_data['Date'] = pd.to_datetime(nse_data['Date'], format='%Y-%m-%d')
nyse_data['Date'] = pd.to_datetime(nyse_data['Date'], format='%Y-%m-%d')

# Display the first few rows to verify the conversion
bse_data.head(5)
```

```python
# Check for missing values
print("BSE Missing Values:")
print(bse_data.isnull().sum())
print("\nNSE Missing Values:")
print(nse_data.isnull().sum())
print("\nNYSE Missing Values:")
print(nyse_data.isnull().sum())
```

```python
# Fill missing values with the mean (if any)
bse_data.fillna(bse_data.mean(), inplace=True)
nse_data.fillna(nse_data.mean(), inplace=True)
nyse_data.fillna(nyse_data.mean(), inplace=True)
```

```python
# No categorical variables to encode in this dataset
# Standardize the numerical features
scaler = StandardScaler()
numerical_features = ['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']

bse_data[numerical_features] = scaler.fit_transform(bse_data[numerical_features])
nse_data[numerical_features] = scaler.fit_transform(nse_data[numerical_features])
nyse_data[numerical_features] = scaler.fit_transform(nyse_data[numerical_features])
```

```python
#Feature Selection
#For simplicity, let's predict whether the closing price is above a certain threshold. This example uses a binary classification based on whether the closing price is above the median value.

# Add a binary target variable: 1 if 'Close' > median, else 0
bse_data['Target'] = (bse_data['Close'] > bse_data['Close'].median()).astype(int)
nse_data['Target'] = (nse_data['Close'] > nse_data['Close'].median()).astype(int)
nyse_data['Target'] = (nyse_data['Close'] > nyse_data['Close'].median()).astype(int)
```

```python
# Select features and target variable
X_bse = bse_data[numerical_features]
y_bse = bse_data['Target']

X_nse = nse_data[numerical_features]
y_nse = nse_data['Target']

X_nyse = nyse_data[numerical_features]
y_nyse = nyse_data['Target']
```

```python
# Split the data into training and testing sets
X_train_bse, X_test_bse, y_train_bse, y_test_bse = train_test_split(X_bse, y_bse, test_size=0.3, random_state=42)
X_train_nse, X_test_nse, y_train_nse, y_test_nse = train_test_split(X_nse, y_nse, test_size=0.3, random_state=42)
X_train_nyse, X_test_nyse, y_train_nyse, y_test_nyse = train_test_split(X_nyse, y_nyse, test_size=0.3, random_state=42)
```

```python
#Model Building and Evaluation

# Logistic Regression Model for BSE Data
model_bse = LogisticRegression()
model_bse.fit(X_train_bse, y_train_bse)
y_pred_bse = model_bse.predict(X_test_bse)

print("BSE Data - Confusion Matrix:")
print(confusion_matrix(y_test_bse, y_pred_bse))
```

```python
print("\nBSE Data - Classification Report:")
print(classification_report(y_test_bse, y_pred_bse))
```

```python
# Logistic Regression Model for NSE Data
model_nse = LogisticRegression()
model_nse.fit(X_train_nse, y_train_nse)
y_pred_nse = model_nse.predict(X_test_nse)

print("NSE Data - Confusion Matrix:")
print(confusion_matrix(y_test_nse, y_pred_nse))
```

```python
print("\nNSE Data - Classification Report:")
print(classification_report(y_test_nse, y_pred_nse))
```

```python
# Logistic Regression Model for NYSE Data
model_nyse = LogisticRegression()
model_nyse.fit(X_train_nyse, y_train_nyse)
y_pred_nyse = model_nyse.predict(X_test_nyse)

print("NYSE Data - Confusion Matrix:")
print(confusion_matrix(y_test_nyse, y_pred_nyse))
```

```python
print("\nNYSE Data - Classification Report:")
print(classification_report(y_test_nyse, y_pred_nyse))
```
