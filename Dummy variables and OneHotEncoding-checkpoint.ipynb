#There are two tupes of data numerical and categorical data and in categorical data there are again two major types nominal(no order between data) and ordinal(Order between data)

<br>#Dummy variables and OneHotEncodinng works on categorical values where there is not order between them, means the data is of nominal categorical data.

```python
import numpy as np
import pandas as pd
```

```python
data = {"Town":["Vapi","Vapi","Vapi","Vapi","Vapi","Anand","Anand","Anand","Anand","Singapore","Singapore","Singapore","Singapore"],
       "Area":[2600,3000,32000,3600,4000,2600,2800,3300,3600,2600,2900,3100,3600], "Price":[550000,565000,610000,680000,725000,
        585000,615000,650000,710000,575000,600000,620000,695000]}
```

```python
#creating a DataFrame
df = pd.DataFrame(data)
df
```

# using the Dummies variable method of pandas

```python
dummies_Town = pd.get_dummies(df["Town"]).astype(int)        #as this was returning boolean value so we use .astype(int)
dummies_Town
```

```python
#Now we are adding or concatenating the dummies_Town and df(DataFrame)
df_dummies = pd.concat([df,dummies_Town],axis="columns")
df_dummies
```

```python
#So, we'll drop the one column, wich we have made by using get_dummies because to avoid dummies variable trap, as it gives us multiple co-linearity between the 
#dimmies columns 
#and we'll drop the Twon column

df_dummies.drop(["Town","Anand"],axis="columns",inplace=True)
df_dummies
```

```python
#now spliitng the DataFrame into X and y
X = df_dummies.drop("Price",axis="columns")
y= df_dummies.Price
```

```python
X
```

```python
y
```

```python
#LinearRegression
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
#predicting the price of area in Anand
model.predict([[3400,0,0]])         #[[]] two brackets beacuse the feature expects 2-D array
#([[area,0,0]]) we wat Anand column price, as it was dropped so 0,0 that means Anand =1
```

```python
#predicting the prce of area in singapore
model.predict([[3800,1,0]])  #[([area,the column we want, 0]])
```

# OneHotEncoding(sklearn)

```python
#creating a new DataFrame dfle from df old DataFrame
dfle=df
dfle
```

```python
#We use LabelEncoder to fit and transform the nominal categorical values into integer values(0,1,2).
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()  #instantiating the object
```

```python
#Using fit_tranform to fit the column Town as an input and tranform the column into integer value.
dfle["Town"] = le.fit_transform(dfle["Town"])
dfle
```

```python
#Now we'll use the OneHotEncoder to make the 0,1 columns like get_dummies of pandas
#we'll use two parameter into the OneHotEncoder, 
#handle_unknown="ignore", if handle_unknown="ignore", then the OneHotEncoder will ignore if there are any unknown values, which were not present during training of data
#sparse_output=False which return the numpy array and which is more redable bu takes more memory and is a pandas DataFrame ,and if Sparse_output=True which is a default which return the sparse matrix which is memory efficient way  of storing large martrices with lots of zeros(0).
#And one function to transform the data into DataFrame
#.set_output(transform="pandas") will  transform the output into a pandas DataFrame
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(handle_unknown="ignore",sparse_output=False).set_output(transform="pandas")        #instantiating the object
```

```python
dfle_Town = ohe.fit_transform(dfle[["Town"]])         #two brackets[[]] because the OneHotEncoder expects the columns as 2-D array
dfle_Town
```

```python
#Now we are concatenating the DataFrame dfle and dfle_Town

dfle = pd.concat([dfle,dfle_Town],axis=1).drop(columns=["Town"])
dfle.head()
```

```python
#Now we are splitting the data into X & y for training
X = dfle.drop(columns=["Price"])
y = dfle["Price"]
```

```python
X.head()
```

```python
#Now we are dropping one column from the new dfle_town(OneHotEncoder) to avoid the dummy variable trap(multiple colinearity between feature columns)
X = X.iloc[:,:3]   #.iloc because we are slicing the DataFrame, if we do't want to use the iloc(implicit slicing) then we can covert the dataframe into
#2_d array values by .values
X
```

```python
y.head()
```

```python
#using the train_test_split model
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
#using the LinearRegression 
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
#getting the best fit line.
lr.fit(X_train,y_train)
```

```python
#predicitng the model
lr.predict([[3400,0,1]])
```
