# <mark><b>Linear regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
np.random.seed(1)  
#np.random.seed() is used so that we can reproduce the same random number numpy array by using the same seed() as 1 into paranthesis
#,as we did in he random_state=1 but we can use the random_state=1 inside the function only but this np.random.seed() function can be used globally, means
#outside the function.
```

```python
num_sample=500    #we have taken 500 number samples
years_of_experience = np.random.randint(2,21,size=num_sample) #so this line of code will generate random integer for 500 samples.
slope = (200_000 - 60_000)/18     #y=m(slope or coefficient)x+c(constant or inercept)
intercept = 60_000
salaries = slope * years_of_experience + intercept + np.random.normal(0,10_000)
#we are using np.random.normal to add noise in the data, because in real life the data is not smooth.
#here np.random.normal(loc,scale,size), is based on the guass's normal distribution, where the .normal() function would take parameters such as,1)loc-which is
#is the mean, where the values would be distributed around the mean ,2) scale-scale is the standard deviation which measures the spread or dispersion in the data,
#the more the standard deviation, more the values would be away from the mean. 3)size=this is the size of the number of sample we want.
```

```python
data={"Years_of_Experience":years_of_experience,"Salary":salaries}
df = pd.DataFrame(data)
print(df)
```

```python
df.head()
```

```python
df.describe()  #.describe() gives us the statistical data and can find out if there are any NAN(not a number) or null values
```

```python
#now we will plot the scatter plot.
plt.figure(figsize=(8,5)) #8inches of width, 5 inches of height
plt.scatter(x="Years_of_Experience",y="Salary",data=df,color="blue")
```

```python
#now we will use the regplot(regression plot), which shows the relationship between the x(independent variable) and y(dependent variable)
sns.regplot(x="Years_of_Experience",y="Salary",data=df,scatter=False,color="red") 
#scatter=True shows the scatteplot along with the regression line
#scatter=False does'nt shows the scatterplot, just shows the regression line
plt.title("regression line")
plt.xlabel("Years_of_Experience")
plt.ylabel("Salary")
plt.show()
```

```python
#now we will train our models, but first we have to split the x and y from the data frame
#here we have given the columns into double brackets because the x is an independent variables means it's in 2-D array.
X=df[["Years_of_Experience"]]  
y=df["Salary"]   #here we have given the y dependent variable in one bracket because it is a dependent variable and it expects 1-D aaray
```

```python
print(X)
```

```python
print(y)
```

```python
#now we will train our model by train_test_split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
lr=LinearRegression() #instantiated the linear regression object
```

```python
lr.fit(X_train,y_train)  #by this we will find the best fit line y=mx+c
```

#.score(), it tells us how accurately out model has predicted the real data, it is know as r2(coefficient of determinant), <mark>if r2=1</mark> it has accurately predicted the actual data, <mark>if r2>0.7</mark> then it shows good accuracy, <mark>if r2 in between 0.5 to 0.7</mark> then it need further analysis and <mark>if r2<0.5</mark> it is poor fit, <mark>if r2<0</mark> model is worse than a horizontal line(i.e., it would be better to predict the mean of the target variables for all observations).

```python
#checking the accuracy of the trained dataset
lr.score(X_train,y_train)
```

```python
#checking the accuracy of the test dataset, how well our model is performing on the unseen data.
#lr.score() gives us the r2(coefficient of determinant)
lr.score(X_test,y_test)
```

```python
#X_test will get multiplied by the coefficient(slope or m) & C and will give us the predicted line y=mx+C
y_pred = lr.predict(X_test)  
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mean_absolute_error(y_pred,y_test)
```

```python
mean_squared_error(y_pred,y_test)
```

```python
r2_score(y_pred,y_test)
```

```python
#now we are checking the values of coefficient and intercept
lr.coef_
```

```python
lr.intercept_
```

```python
coefficients = lr.coef_
intercept = lr.intercept_
```

```python
X=np.linspace(0,20,100)
#np.linspace will create a linearly spaced array where 0 is the start,20 is the end of an array and 100 is number of samples(create 100 number of samples from 0 to 20).
y=coefficients * X + intercept     #y=mx+c
```

```python
#coefficient(m) is univarient & here we have only one feature(one independent varable/dimension), so we have to mention the index of the slope we want.
#label() is used to show the legend in the plot, and to show it we have to write plt.legend()
plt.scatter(X,y,label=f"y={coefficients[0]:.2f} X + {intercept:.2f}", color="blue")   #coefficients[0]:.2f means, give me the coefficients at 0th index till 2 decimals
plt.title("Linear Regression")
plt.xlabel("X-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```

# <mark><b>Multi-variant Linear Regression Model

```python
#Multiple variable or feature
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

```python
#here we are building a model so we can figure out the wins of a baseball teams, interms of their hits, homeruns, losses, triples,strikeouts, etc.
```

```python
#creating a dataframe.
data = []
for _ in range(500):  #here we have used a blank(underscore) because we don't want to use the i(iterator) in the function.so, we can use it like this.

    team_name = f"Team{chr(random.randint(65,90))}"  #here we have imported random so we don't have to use numpy(np)
    # random.randint will generate a random integer from 65 to 90 and will be converted by chr into a uppercase charater from A to Z and will return a team
    #name like Team A, Team U, etc
    seasons = random.randint(2012,2024)
    wins = random.randint(50,110)
    losses = 162 - wins
    hits = random.randint(1200,1600)
    doubles = random.randint(200,350)
    triples = random.randint(10,40)
    home_runs = random.randint(100,250)
    strikeouts = random.randint(1000,1500)

    #print(f"iterations{i}:Team name:{team_name},Seasons:{seasons},Wins:{wins},Losses:{losses} ,Hits:{hits}, Doubles:{doubles}, Triples:{triples}, Home_runs:home_runs, Strikeouts:{strikeouts}")
    #team_data = {
     #   "team_name": team_name,
      #  "season": seasons,
       # "wins": wins,
        #"losses": losses,
        #"hits": hits,
        #"doubles": doubles,
        #"triples": triples,
        #"home_runs": home_runs,
        #"strikeouts": strikeouts
    #}
    
    #data.append(team_data)
    
    #here now we will adjust the above data(variables), so that we don't get much variability(randomness).
    hits_adjusted = hits + (wins-80) * 5
    doubles_adjusted = doubles + (wins-80) * 2
    triples_adjusted = triples + (wins -80)
    home_runs_adjusted = home_runs + (wins-80) * 3
    strikeouts_adjusted = strikeouts - (wins-80) * 10
    
    #now appending the data in to the empty variable data=[]
    data.append([team_name, seasons,wins,losses,hits_adjusted,doubles_adjusted,triples_adjusted,home_runs_adjusted,strikeouts_adjusted])
    
```

```python
#now we are giving names to the columns
columns = ["Teams","Seasons","Wins","Losses","Hits","Doubles","Triples","Homeruns","Strikeouts"]
```

```python
df = pd.DataFrame(data, columns=columns)
df.head()
```

```python
#plotting the data 
import matplotlib.pyplot as plt
import seaborn as sns
sns.lmplot(x="Hits",y="Wins",data=df)    #lmplot is the linear model plot which shows the linear relationship between x and y
plt.show()
```

```python
sns.regplot(x="Hits",y="Wins",data=df)    #lmplot is the linear model plot which shows the linear relationship between x and y
#(corelation coefficient(r)=1)
plt.show()
```

```python
#this was the positive linear relationship, now we will be seeing the negative linear relationship graph.(corelation coefficient(r)=-1)
#this sns.lmplot() is a linear model plot where it shows the relationship between x(feature variable) and y (target variable)
sns.lmplot(x="Strikeouts",y="Wins",data=df)
plt.show()
```

```python
#now we will drop some of the columns such as Teams(as Teams is a categorical data & the target column would'nt have much impact by Teams), Losses & Seasons.
df2 = df.drop(columns=["Teams","Losses","Seasons"])
df2.head()
```

```python
#now we are splitting the data into X and y.
X = df2[["Hits","Doubles","Triples","Homeruns","Strikeouts"]]   #two brackets because the features expects 2-D arrray because it can have multiple columns/dimensions
y = df2["Wins"]   #one bracket because it expects 1-D array because in target, it only has one column/dimension.
```

```python
#importing the LinearRegreesion() model
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
#now we are training the data through train_test_split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
#finding the best fit line(by m & c)
lr.fit(X_train,y_train)
```

```python
#finding the r2(coefficient of determinant) on training data.
lr.score(X_train,y_train)
```

```python
#finding the r2(coefficient of determinant) on test data
lr.score(X_test,y_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
#we have best fit line which we have got from the lr.fit(X_train,y_train), now we will input the X_test data into this best fit line and then m(coefficient or slope)
# will get multiplied with X_test and will give us the y predction.
y_pred = lr.predict(X_test)
```

```python
#now we will check teh accuraccy of the y_pred with the actual y(y_test)
mean_absolute_error(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```

```python
#now we will check teh coefficients and intercept
lr.coef_
```

```python
lr.intercept_
```

```python
coefficients = lr.coef_
intercept = lr.intercept_
n_features = 5
#linspace is creating 100 integers from 0 to 20 in linear way
#reshape(-1, 1) is a commonly used operation in NumPy to reshape an array. Here's what it does:
#-1: This tells NumPy to automatically calculate the size of this dimension based on the total size of the array and the other specified dimensions.
#It essentially means "as many rows as necessary."
#1: This specifies that the resulting array should have 1 column.
new_df  = np.linspace(0,20,100).reshape(-1,1)
mult_df = np.hstack([new_df] * n_features)

y = np.dot(mult_df, coefficients) + intercept
plt.plot(new_df, y, label='Regression Line')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.legend()
plt.show()
```
