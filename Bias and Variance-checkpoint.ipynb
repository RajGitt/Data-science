<b>Overfit</b> :- when the decision boundary is not a smooth curve then the model is overfit, and the model would be <mark>overfit</mark> when the decision boundary is passing through the all the training data, 
<br>so that the training error would be zero but there would be test data error. And whenever we are choosing the sample training data randomly there would be some test data error, so the difference in test data error is known as variance if the test data error on one random training of data is something then the test error for the second random training of data is higher then the variance is <b>high variance</b> and if the difference is low then the variance is <b>low variance</b>.

<b>Underfit</b> :- when the decision boundary is a (straight line)linear equation then the model would be <mark>underfit</mark> model, as the decision boundary is not passing through most of the data point of the training data, leaving some data from th etraining data and test data. So, there would be train data error and test data error, if the difference of the one random training data is high then the other training data by random training of data then the bias would be <b>high bias</b> and if the training data of one training data is low then the other training data by random training of data the the bias is <b>low bias</b>

<b>Balancedfit</b> :- The model is said to be balanced fit when the variance and bias in the data of the model is low.
